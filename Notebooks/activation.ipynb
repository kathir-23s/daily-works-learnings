```python
import matplotlib.pyplot as plt
import math
import time
```

```python
inputs = [i * 0.1 for i in range(-50000, 50001)] 
```

```python
def sigmoid(x):
	try:
		z = 1 / (1 + math.exp(-x))
		return z
	except OverflowError:
		return float("inf")

def sigmoid_derivative(x):
				z = sigmoid(x) * (1 - sigmoid(x))
				# z = math.exp(x) / (1 + (math.exp(x)))**2
				return z
```

```python
sigmoid_a = []
sigmoid_deriv = []
start = time.time()
for i in inputs:
				a, da = sigmoid(i) , sigmoid_derivative(i)
				sigmoid_a.append(a)
				sigmoid_deriv.append(da)
f'{time.time() - start:.8f} sec'
```

```python
plt.xlim(-50000, 50001)
plt.plot(inputs, sigmoid_a, color="green", label="Sigmoid of x")
plt.plot(inputs, sigmoid_deriv, color="red", linestyle="--", label="Derivative of sigmoid of x")

plt.xlabel("inputs")
plt.ylabel("\u03c3(x) and \u03c3\'(x)")

plt.legend()
```
