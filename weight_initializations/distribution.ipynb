```python
import numpy as np
import matplotlib.pyplot as plt

# Generate normally distributed data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=10000)  # mean=0, std=1

# Apply tanh transformation
tanh_data = np.tanh(data)

# Derivative of tanh
x_vals = np.linspace(-4, 4, 400)
tanh_derivative = 1 - np.tanh(x_vals)**2

# --- Metric: values outside (-1, 1) in Gaussian ---
outside_mask = (data < -1) | (data > 1)
outside_count = np.sum(outside_mask)
outside_fraction = np.mean(outside_mask)

# Their mapped tanh values
outside_tanh = tanh_data[outside_mask]

print(f"Count outside [-1,1] in Gaussian: {outside_count}")
print(f"Fraction outside [-1,1]: {outside_fraction:.2%}")
print(f"After tanh, these values cluster near edges:")
print(f"  Mean = {outside_tanh.mean():.3f},  Min = {outside_tanh.min():.3f},  Max = {outside_tanh.max():.3f}")

# --- Visualization ---
plt.figure(figsize=(18,6))

# Original Gaussian
plt.subplot(1,3,1)
plt.hist(data, bins=50, density=True, alpha=0.7, color='blue', label="All Data")
plt.hist(data[outside_mask], bins=50, density=True, alpha=0.6, color='red', label="Outside [-1,1]")
plt.axvline(-1, color='black', linestyle='--')
plt.axvline(1, color='black', linestyle='--')
plt.title("Normal Distribution (Bell Curve)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()

# After tanh
plt.subplot(1,3,2)
plt.hist(tanh_data, bins=50, density=True, alpha=0.7, color='green', label="All Data")
plt.hist(outside_tanh, bins=50, density=True, alpha=0.6, color='orange', label="Mapped Tail Values")
plt.title("After Applying tanh()")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()

# Derivative of tanh
plt.subplot(1,3,3)
plt.plot(x_vals, tanh_derivative, color='purple', linewidth=2)
plt.title("Derivative of tanh(x)")
plt.xlabel("x")
plt.ylabel("f'(x) = 1 - tanh(x)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# 1. Effect of Mean
means = [-2, 0, 2]
std = 2
mean_styles = [('-', 'red'), ('--', 'orange'), (':', 'magenta')]
for m, (ls, color) in zip(means, mean_styles):
    axs[0].plot(x, 1/(std*np.sqrt(2*np.pi)) * np.exp(-(x-m)**2/(2*std**2)),
                linestyle=ls, color=color, linewidth=2, label=f"mean={m}")
axs[0].set_title("Effect of Mean (std=2)")
axs[0].legend()
axs[0].grid(True)

# 2. Effect of Standard Deviation
mean = 0
stds = [1, 2, 3]
std_styles = [('-.', 'blue'), ((0, (5, 5)), 'cyan'), ((0, (1, 10)), 'navy')]
for s, (ls, color) in zip(stds, std_styles):
    axs[1].plot(x, 1/(s*np.sqrt(2*np.pi)) * np.exp(-(x-mean)**2/(2*s**2)),
                linestyle=ls, color=color, linewidth=2, label=f"std={s}")
axs[1].set_title("Effect of Std (mean=0)")
axs[1].legend()
axs[1].grid(True)

# 3. Effect of Variance
vars = [1, 4, 9]
var_styles = [((0, (3, 1)), 'green'), ((0, (1, 1)), 'lime'), ((0, (3, 5, 1, 5)), 'darkgreen')]
for v, (ls, color) in zip(vars, var_styles):
    s = np.sqrt(v)
    axs[2].plot(x, 1/(s*np.sqrt(2*np.pi)) * np.exp(-(x-mean)**2/(2*s**2)),
                linestyle=ls, color=color, linewidth=2, label=f"var={v}")
axs[2].set_title("Effect of Variance (mean=0)")
axs[2].legend()
```
